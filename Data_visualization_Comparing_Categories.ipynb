```python
import pandas as pd
import os
csv_directory = "/content/"

combined_df = pd.DataFrame()

# Columns to append to the combined DataFrame
columns_to_append = ['common_name', 'scientific_name', 'city', 'height_M']

# Loop through the files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(csv_directory, filename)

        # Read each CSV file into a temporary DataFrame
        temp_df = pd.read_csv(file_path, error_bad_lines=False)

        # Check if all the specified columns are present in the CSV file
        if all(col in temp_df.columns for col in columns_to_append):
            # Drop rows with missing values in any of the specified columns
            temp_df = pd.read_csv(file_path, usecols=columns_to_append, error_bad_lines=False)
            temp_df.dropna(subset=columns_to_append, inplace=True)

            # Append the data from the temporary DataFrame to the combined DataFrame
            combined_df = combined_df.append(temp_df, ignore_index=True)

# Group the data by 'scientific_name' and count the number of trees for each species
species_count = combined_df['scientific_name'].value_counts()
```

```python

def calculate_species_counts(city, dataframe, top_species):
    scientific_name = 'scientific_name'

    # Create a dictionary to store the counts
    value_counts = {value: (dataframe[scientific_name] == value).sum() for value in top_species}

    species_to_count = []
    new_row = {'city': city}

    total_count = sum(value_counts.values())  # Calculate the total count for the city

    # Print the counts and add them to the new_row dictionary
    for value, count in value_counts.items():
        species_to_count.append(value)
        new_row[value] = count

    new_row['total'] = total_count  # Add the total count to the new_row dictionary

    # Create a DataFrame with the specified columns
    new_columns = ['city', 'scientific_name']
    new_df = dataframe[new_columns]

    # Append the new_row to the result_df
    result_df = pd.DataFrame(columns=['city'] + species_to_count + ['total'])
    result_df = result_df.append(new_row, ignore_index=True)

    return result_df

# Define the top species and dataframes for both cities
top_species = combined_df['scientific_name'].value_counts().head(8).index.tolist()
# Create the dataframes df1 and df2 for San Diego and Oakland, respectively

# Read the first CSV file into a DataFrame
csv_file1 = "/content/Tampa_Final_2022-06-18.csv"
df1 = pd.read_csv(csv_file1)

# Read the second CSV file into a DataFrame
csv_file2 = "/content/Oakland_Final_2022-06-18.csv"  # Replace with the path to your second CSV file
df2 = pd.read_csv(csv_file2)

# Read the first CSV file into a DataFrame
csv_file3 = "/content/Orlando_Final_2022-06-18.csv"
df3 = pd.read_csv(csv_file3)

# Read the second CSV file into a DataFrame
csv_file4 = "/content/RanchoCucamonga_Final_2022-06-18.csv"  # Replace with the path to your second CSV file
df4 = pd.read_csv(csv_file4)

# Calculate species counts for San Diego
result_df1 = calculate_species_counts('San Diego', df1, top_species)

# Calculate species counts for Oakland
result_df2 = calculate_species_counts('Oakland', df2, top_species)

# Calculate species counts for San Diego
result_df3 = calculate_species_counts('Orlando', df3, top_species)

# Calculate species counts for Oakland
result_df4 = calculate_species_counts('Rancho Cucamonga', df4, top_species)

# Combine the results if needed
final_result = pd.concat([result_df1, result_df2, result_df3, result_df4], ignore_index=True)

# Print the final result
final_result
```
